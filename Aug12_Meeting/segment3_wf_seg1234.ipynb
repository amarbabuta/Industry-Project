```python
for i in segment3_df['CASE_ID'].unique():
    if i in caseid_w1 : 
        caseid_length = len(segment3_df[segment3_df['CASE_ID'] == i])
        info = segment3_df[segment3_df['CASE_ID'] == i]
        workflow_segment1 = workflow_segment1.append(info)
```

```python
workflow_segment1
```

```python
wf_sg1_list = dict()
for i in range(workflow_segment1.shape[0]):
    for j in workflow_segment1['CASE_ID'].unique():
        if j == workflow_segment1['CASE_ID'].iloc[i]:
            if j not in wf_sg1_list:
                if workflow_segment1['WRKFLW_TSK_MAP_GROUPED'].iloc[i] == "Send for Assessment":
                    flag1 = i
                    time1 = workflow_segment1["CreateTime"].iloc[i]
                    wf_sg1_list[j] = [flag1, {"SAF":time1}]
            else:
                if workflow_segment1['WRKFLW_TSK_MAP_GROUPED'].iloc[i] == "Submit":
                    flag2 = i
                    time2 = workflow_segment1["CreateTime"].iloc[i]
                    wf_sg1_list[j] += [flag2, {"Submit":time2}]
```

```python
print(len(wf_sg1_list))
```

```python
ws1 = pd.DataFrame(columns = ["case_id","number of cases","processing_time"])
```

```python
for k,v in wf_sg1_list.items():
    number = v[2]-v[0]
    for key1,value1 in v[1].items():
        time1 = pd.to_datetime(value1)
    for key2,value2 in v[3].items():
        time2 = pd.to_datetime(value2)
    processtime = time1-time2
    info = {"case_id":k, "number of cases":number, "processing_time":processtime}
    ws1 = ws1.append(info, ignore_index=True)
```

```python
ws1
```

```python
ws1["number of cases"].unique()
```

```python
time1_list = dict()
for i in range(ws1.shape[0]-1):
    for j in ws1["number of cases"].unique():
        if j == ws1["number of cases"].iloc[i]:
            if j not in time1_list:
                time = ws1["processing_time"].iloc[i]
                time1_list[j] = [time]
            else:
                time = ws1["processing_time"].iloc[i]
                time1_list[j] += [time]
```

```python
workflow_seg1_list = pd.DataFrame(columns = ["number", "number of cases","total_processing_time", "average_processing_time", "% of cases reaching to end of workflow segment"])
```

```python
for k,v in time1_list.items():
    time_length = len(v)
    for i in range(time_length):
        if i == 0:
            sum_time = pd.Timedelta(v[i])
            count = 1
        else:
            sum_time += pd.Timedelta(v[i])
            count += 1
    avg_processing_time = sum_time/count
    percentage = '{:.4f}%'.format(count/647*100)
    info = {"number":k, "number of cases":count, "total_processing_time":sum_time, "average_processing_time":avg_processing_time, "% of cases reaching to end of workflow segment":percentage}
    workflow_seg1_list = workflow_seg1_list.append(info, ignore_index=True)
```

```python
workflow_seg1_list.sort_values("number",inplace=True)
```

#### Workflow Segment2
Possible matrix to help identify blockages in each of segment 1, 2, 3 = HDCW students + Engineering + OF/AS/AF <br>
Workflow Segment 2 - Send for Assessment to Assessment Outcome Successful/Unsuccessful <br>
1. Number of cases<br>
2. average time for workflow segment<br>
3. median time for workflow segment<br>
4. % of cases reaching to end of workflow segment<br>

```python
"""
workflow segment1234:
Submit -- Send for Assessment -- Assessment Outcome Successful/Unsucessful -- Release Offer -- Accept/Defer, 
it can be regarded workflow_segment1 data works for four workflow segments studies
"""
workflow_segment2 = workflow_segment1.copy()
wf_sg2_list = dict()
for i in range(workflow_segment2.shape[0]):
    for j in workflow_segment2['CASE_ID'].unique():
        if j == workflow_segment2['CASE_ID'].iloc[i]:
            if j not in wf_sg2_list:
                if workflow_segment2['WRKFLW_TSK_MAP_GROUPED'].iloc[i] == "Assessment Outcome - Successful":
                    flag1 = i
                    time1 = workflow_segment2["CreateTime"].iloc[i]
                    wf_sg2_list[j] = [flag1, {"S":time1}]
                elif workflow_segment2['WRKFLW_TSK_MAP_GROUPED'].iloc[i] == "Assessment Outcome - Not Successful":
                    flag2 = i
                    time2 = workflow_segment2["CreateTime"].iloc[i]
                    wf_sg2_list[j] = [flag2, {"NS":time2}]
            else:
                if workflow_segment2['WRKFLW_TSK_MAP_GROUPED'].iloc[i] == "Send for Assessment":
                    flag3 = i
                    time3 = workflow_segment2["CreateTime"].iloc[i]
                    wf_sg2_list[j] += [flag3, {"SFA":time3}]
```

```python
# in workflow segment2 , it contains 390 unqiue caseid
print(len(wf_sg2_list))
```

```python
"""
Notice :
1. in wf_sg2_list, remove those records which contain wrong orders between two workflows (not sure)
   in this case, in origin dataset(PG.csv), for a unique caseid :
           row[i] (Send for Accessment)
           row[i+1] (Assessment Outcome - Successful/Unsuccessful)
           
2. other special case : in origin dataset(PG.csv), for a unique caseid :
            row[i] (Assessment Outcome - Successful/Unsuccessful)
            row[i+1] (Send for Accessment)
            row[i+2] (Send for Accessment)
    here calculate processing time as row[i].time(i) - row[i+1].time(i+1)
    
"""
```

```python
ws2 = pd.DataFrame(columns = ["case_id","number of cases","processing_time"])
```

```python
for k,v in wf_sg2_list.items():
    length = len(v)
    if length > 2:
        number = v[2]-v[0]
        for key1,value1 in v[1].items():
            time1 = pd.to_datetime(value1)
        for key2,value2 in v[3].items():
            time2 = pd.to_datetime(value2)
        processtime = time1-time2
        info = {"case_id":k, "number of cases":number, "processing_time":processtime}
        ws2 = ws2.append(info, ignore_index=True)
```

```python
# After filtering data, workflow segment2 contains 370 valid unique caseid
ws2
```

```python
time2_list = dict()
for i in range(ws2.shape[0]-1):
    for j in ws2["number of cases"].unique():
        if j == ws2["number of cases"].iloc[i]:
            if j not in time2_list:
                time = ws2["processing_time"].iloc[i]
                time2_list[j] = [time]
            else:
                time = ws2["processing_time"].iloc[i]
                time2_list[j] += [time]
```

```python
workflow_seg2_list = pd.DataFrame(columns = ["number", "number of cases","total_processing_time", "average_processing_time", "% of cases reaching to end of workflow segment"])
```

```python
for k,v in time2_list.items():
    time_length = len(v)
    for i in range(time_length):
        if i == 0:
            sum_time = pd.Timedelta(v[i])
            count = 1
        else:
            sum_time += pd.Timedelta(v[i])
            count += 1
    avg_processing_time = sum_time/count
    percentage = '{:.4f}%'.format(count/370*100)
    info = {"number":k, "number of cases":count, "total_processing_time":sum_time, "average_processing_time":avg_processing_time, "% of cases reaching to end of workflow segment":percentage}
    workflow_seg2_list = workflow_seg2_list.append(info, ignore_index=True)
```
