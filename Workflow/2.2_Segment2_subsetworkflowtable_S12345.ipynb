```python
for i in range(len(df2)):
    id=df2.loc[i,'CASE_ID']
    status = df2.loc[i,"WRKFLW_TSK_MAP_GROUPED"]
```

```python
ids = list(set(df2['CASE_ID'].values.tolist()))
ids
```

```python
S2S1 = pd.DataFrame()
S2S2 = pd.DataFrame()
S2S3 = pd.DataFrame()
S2S4 = pd.DataFrame()
S2S5 = pd.DataFrame()

#traverse each id
for id in ids:
    flag = True
    id_df = df2[df2['CASE_ID'] == id].loc[::-1].reset_index(drop = True)
    #id_df= id_df.iloc[::-1]
    sub_index,sendForAssessment_index,assessmentOutcomeSorNotS_index,releaseOffer_index,acceptOrDefer_index = -1,-1,-1,-1,-1
    #.loc[::-1].reset_index(drop = True) Reverse row order and reset index which let the first status stay at the first row:
    for index, row in id_df.iterrows():
        #record the index of first 'Submit' status per person
        if row['WRKFLW_TSK_MAP_GROUPED'] == 'Submit'and sub_index == -1:
            sub_index = index
        #record the index of first 'Send for Assessment' status per person
        elif row['WRKFLW_TSK_MAP_GROUPED'] == 'Send for Assessment'and sendForAssessment_index == -1:
            sendForAssessment_index = index
        #record the index of first 'Assessment Outcome' status per person
        elif (row['WRKFLW_TSK_MAP_GROUPED'] == 'Assessment Outcome - Successful'and assessmentOutcomeSorNotS_index == -1) \
                or (row['WRKFLW_TSK_MAP_GROUPED'] == 'Assessment Outcome - Not Successful'and assessmentOutcomeSorNotS_index == -1):
            assessmentOutcomeSorNotS_index = index
        #record the index of first 'Release Offer' status per person
        elif row['WRKFLW_TSK_MAP_GROUPED'] == 'Release Offer' and releaseOffer_index == -1:
            releaseOffer_index = index
        #record the index of first 'Accept/Defer' status per person
        elif (row['WRKFLW_TSK_MAP_GROUPED'] == 'Accept' and acceptOrDefer_index == -1) or\
                (row['WRKFLW_TSK_MAP_GROUPED'] == 'Defer' and acceptOrDefer_index == -1):
            acceptOrDefer_index = index
    #save the person status between submit and sendforassessment into S1
    if sub_index != -1 and sendForAssessment_index != -1:
        df_seg1 = id_df[sub_index:sendForAssessment_index+1]
        S2S1 = pd.concat([S2S1,df_seg1])
        flag = False
    #save the person status from submit and before sendforassessment into S1
    elif sub_index != -1 and sendForAssessment_index == -1 and assessmentOutcomeSorNotS_index == -1\
        and releaseOffer_index == -1 and acceptOrDefer_index == -1:
        df_seg1 = id_df
        S2S1 = pd.concat([S2S1,df_seg1])
        flag = False
    #save the person status between sendforassessment and assessmentOutcome into S2
    if sendForAssessment_index != -1 and assessmentOutcomeSorNotS_index != -1:
        df_seg2=id_df[sendForAssessment_index:assessmentOutcomeSorNotS_index+1]
        S2S2 = pd.concat([S2S2,df_seg2])
        flag = False
    #save the person status from sendforassessment and before assessmentOutcome into S2
    elif sendForAssessment_index != -1 and assessmentOutcomeSorNotS_index == -1\
        and releaseOffer_index == -1 and acceptOrDefer_index == -1:
        df_seg2=id_df[sendForAssessment_index:]
        S2S2 = pd.concat([S2S2,df_seg2])
        flag = False
    #save the person status between assessmentOutcome and relaseoffer into S3
    if assessmentOutcomeSorNotS_index!=-1 and releaseOffer_index != -1:
        df_seg3=id_df[assessmentOutcomeSorNotS_index:releaseOffer_index +1]
        S2S3 = pd.concat([S2S3,df_seg3])
        flag = False
    #save the person status from assessmentOutcome and before relaseoffer into S3
    elif assessmentOutcomeSorNotS_index != -1 and releaseOffer_index == -1\
            and acceptOrDefer_index == -1:
        df_seg3=id_df[assessmentOutcomeSorNotS_index:]
        S2S3 = pd.concat([S2S3,df_seg3])
        flag = False
    #save the person status between relaseoffer and acceptordefer into S4
    if releaseOffer_index!=-1 and acceptOrDefer_index != -1:
        df_seg4=id_df[releaseOffer_index:acceptOrDefer_index+1]
        S2S4 = pd.concat([S2S4,df_seg4])
        flag = False
    #save the person status from relaseoffer and before acceptordefer into S4
    elif releaseOffer_index != -1 and acceptOrDefer_index == -1:
        df_seg4=id_df[releaseOffer_index:]
        S2S4 = pd.concat([S2S4,df_seg4])
        flag = False
    #save the person who didn't belong to before four segements
    if flag == True:
        S2S5 = pd.concat([S2S5,id_df])
```
